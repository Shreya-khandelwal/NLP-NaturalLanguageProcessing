```python
#import library and language model
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(u"The quick brown fox jumped over the lazy dog's back.")
```

```python
# import displacy library
from spacy import displacy
```

```python
#render the dependency and visualize it using displacy render
displacy.render(doc, style='dep')
```

```python
#options is a optional argument
#customise the appearance
options = {'distance':90, 'compact':True, 'color':'yellow', 'bg':'purple', 'font':'Times'}
```

```python
displacy.render(doc, style='dep', options=options)
```

```python
doc_2 = nlp(u"This is a sentence. This is another, possibly longer sentence.")
```

```python
#splittng sentences in a document
spans = list(doc_2.sents)
```

```python
#render this on the external server using displacy serve
displacy.serve(spans, style='dep', options={'distance':200})
```
